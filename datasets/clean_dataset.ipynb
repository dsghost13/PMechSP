```python
import csv

INPUT_CSV = '10k_Proton_Transfer_OG.csv'
OUTPUT_CSV = '10k_Proton_Transfer.csv'

with open(INPUT_CSV, newline='', encoding='utf-8') as csv_in, \
     open(OUTPUT_CSV, 'w', newline='', encoding='utf-8') as csv_out:

    reader = csv.reader(csv_in)
    writer = csv.writer(csv_out)

    for row in reader:
        if row:
            # keep reactants only
            row[0] = row[0].split('>>')[0]
        writer.writerow(row)
```

```python
import pandas as pd
import re

# Load CSV
df = pd.read_csv("13k_All_Manual.csv")

def contains_only_20_21(cell):
    # If the cell is a list, flatten it to a string
    if isinstance(cell, list):
        cell_str = " ".join(map(str, cell))
    else:
        cell_str = str(cell)
    # Extract all numbers
    numbers = re.findall(r'\b\d+\b', cell_str)
    # Return True if at least one number exists and all numbers are 20 or 21
    return bool(numbers) and all(n in ('20', '21') for n in numbers)

# Apply function safely
mask = df["SMILES Labelled"].apply(contains_only_20_21)

# Use the boolean mask to filter
filtered_df = df[mask]

# Save result
filtered_df.to_csv("test.csv", index=False)

print(f"Filtered {len(filtered_df)} rows containing only 20 and 21 as numbers.")
```

```python
import csv

INPUT_FILES = [
    '4k_Clayden.csv',
    '10k_Mayr.csv',
    '10k_Proton_Transfer.csv',
    '13k_All_Manual.csv'
]

OUTPUT_FILE = '37k_Combined.csv'

with open(OUTPUT_FILE, 'w', newline='', encoding='utf-8') as csv_out:
    writer = csv.writer(csv_out)
    header_written = False

    for file in INPUT_FILES:
        with open(file, newline='', encoding='utf-8') as csv_in:
            reader = csv.reader(csv_in)
            header = next(reader, None)

            # write header only once
            if not header_written and header:
                writer.writerow(header)
                header_written = True

            # write remaining rows
            for row in reader:
                if row:  
                    writer.writerow(row)

print("Finished combining!")
```
